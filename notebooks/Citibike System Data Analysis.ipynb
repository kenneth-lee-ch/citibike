### ii. Unsupervised Learning - Clustering


* Next, we want to understand how to better allocate our bikes in different bike stations and gain insights to improve citi bike services. Our approach is to cluster all the bike stations by using their geographic information and their corresponding bike pick-up and return frequencies on average in each hour of a day. 

* The reason behind this idea is that we think there may be some bike stations located in downtown business areas where a lot of companies, industries and  headquarters generate numerous employment may receive high and concentrated volume of bike returns in the weekday morning peak period. Also, the bike stations located around the subway lines may encounter intensive pickup demands at a certain time whereas the capacity of these bike stations can not support. On the contrary, a large portion of bikes in the bike sharing system may not be in efficient use and thus face a long time with empty occupancy. 

* To evluate our clustering techniques, we make use of Silhouette score and Davies-Bouldin score, which evaluate the cohesion and separation of data points respectively. They can give us a succinct graphical representation of how well each cluster has been classified. 

**Warning:** The following one cell may take half an hour to run as multiple queries will be made. 

### Summary for Clustering Procedures
- As a summary, there are 69 variables in the extracted dataset for clustering, which are standardized for the ongoing cluster analysis. 
- We are trying three different clustering algorithms and to make decision on which algorithm and the number of clusters to use finally. 
- The metric we adopt to measure the performance is SIlhouette Index and Davies-Bouldin Index, which represent the cohesion and seperation of data points. 
- k-means, Birch (balanced iterative reducing and clustering using hierarchies) and Spectral Biclustering (of data dimension reduction methods) are applied for cluster analysis. K-Means is one of the most well-known partitional cluster algorithms.

```python
#generate series of hours from 0 to 24
hour = ["%.2d" % i for i in range(25)]
#initialize empty dataframe for merging
#O for origin(pickup), D for destination(return)
OD = pd.DataFrame(columns = ['Station','ID'])
# obtain mean durations for p/r and number of p/r by hour and combine it to a complete dataframe
for i in range (24):
    # Call the method from the data_preprocess.py
    OD_hour = get_OD(hour[i],hour[i+1])
    OD = pd.merge(OD,OD_hour,on=['Station','ID'],how='outer')

##clean the OD data
OD = clean_OD(OD)
```

```python
##calculate fration of pickups/returns
temporal = OD[['ID']]
#get the last 4 columns of variables at the beginning of dataframe
temporal = temporal.merge(OD.iloc[:,-4:],left_index=True,right_index=True)
#calculate fraction of pickups/returns by hour
for i in range(0,24):
    temporal['pickup'+str(i)] = OD.iloc[:,3+i*4]/temporal['pickup_sum']
    temporal['return'+str(i)] = OD.iloc[:,5+i*4]/temporal['return_sum']
temporal

#load the spatial data extracted by ArcGIS (external)
spatial = pd.read_csv('data/spatial.csv').drop_duplicates('ID')

#combine into final dataset for clustering
cluster = temporal.merge(right=spatial,on='ID',how='inner')
cluster
```

**Standardization & weighting of variables**

```python
##standardization of variables
X = standardize(cluster.iloc[:,1:])  #neglect the first ID column

##giving weights for temporal and spatial variables 
##for weight balancing (too many temporal variables)

#first part
X1 = X.iloc[:,0:2]*4 #biking activities
#second part
X2 = X.iloc[:,2:4]*3.5 #mean duration
#third part
X3 = X.iloc[:,4:52]*1 #temporal
#forth part
X4 = X.iloc[:,52:]*5 #spatial

#combine to weighted data
Y = X1.merge(right=X2,left_index=True,right_index=True)\
.merge(right=X3,left_index=True,right_index=True)\
.merge(right=X4,left_index=True,right_index=True)
Y
```

**Silhouette Score vs. Davies Bouldin Score**

```python
#plot results of Silhouette score
plt.rc('font', family='arial', size=12)
cluster_validation_sil(Y).iloc[:,:3].plot(kind='line',linewidth=2,
                                   alpha=0.7,rot=1,figsize=(5,4),legend=True)
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
#plt.yticks(np.arange(0,10000,2000))
plt.xticks(np.arange(0,8,1),np.arange(3,11,1))
plt.grid(axis="y",linestyle='--',linewidth=0.5)
plt.show()
```

**Figure 2-2 Evaluation based on Davies Bouldin Score to find the number of cluster**: 

Figure 2-1 and 2-2 show the Silhouette & Davies-Bouldin scores corresponding to the number of clusters by implementing three different clustering algorithms.
- The higher the value for Silhouette Index the better the clustering. The opposite holds for the Davies-Bouldin-Index.
- A first (local) optimum in the indices' values, called _elbow_, indicates a reasonable clustering. 
- A clear elbow indicated by a drop in the Davies-Bouldin-Index and peaks for Silhouette can be identified for the K-Means with 5 clusters. 
- We choose the __number of clusters to be 5__ and use __K-Means__ algorithm to do the following clustering.

**K-Means Clustering**

**Warning**: 
The following analysis is written based on the order of clusters. The order may change if the following cell is re-run. 

```python
#clustering into 5 clusters
c = KMeans(n_clusters=5,random_state=78).fit(Y)
print(silhouette_score(Y, c.labels_))

#append cluster label result to original data
result = cluster
result['label'] = c.labels_
#rename clusters
result['label'] = result['label'].replace({0:'Class 1',
                                  1:'Class 2',
                                  2:'Class 3',
                                  3:'Class 4',
                                  4:'Class 5'})
#read in station list with lat and lon which is saved by making query in SQL in the previous part
station_list = station_list.drop_duplicates('ID')
#merge cluster result with lat and lon of stations
result = pd.merge(result, station_list, on=['ID'], how='left')
group_result = result.groupby('label')
print(group_result.size())
```

The K-Means clustering algorithm for five clusters is finally implemented by the indication of Silhouette and Davies-Bouldin scores and thus five clusters of bike stations in New York city are obtained. Regarding the activity at stations, the mean values which can be conceived as the centroids of each cluster can be validated both temporally and spatially. In a summary, the numbers of stations assigned to each cluster are 261, 121, 345, 155 and 50, from Class 1 to Class 5 clusters respectively. The consequent sample sizes of clusters are preferable, which also meet the prespecified clustering examination criteria.

```python
def plot_cluster(cluster, color):
    """
    plot the hourly fractions of pickups and returns of each cluster
    Arguments: 
        cluster: name of the cluster predefined, str
        color: six-digit color series name, str 
    """
    plt.rc('font', family='arial', size=12)
    _pickup[cluster].plot(kind='line',linestyle='-',color=color, label=cluster + ' pickup')
    _return[cluster].plot(kind='line',linestyle='--',color=color, label=cluster + ' return')
    plt.xlabel('Hour of Day')
    plt.ylabel('Fraction of Daily Activities (%)')
    plt.legend()
    plt.yticks(np.arange(0,0.2,0.04),np.arange(0,20.0,4.0))
    plt.xticks(np.arange(0,30,6),np.arange(0,30,6))
    plt.grid(axis="y",linestyle='--',linewidth=0.5)
    return
```

```python
fig=plt.figure(figsize=(17,8))
columns = 3
rows = 2
colors = ['#3498db','#1abc9c','#f1c40f','#e67e22','#e74c3c']

plt.title('Pickups & Returns in the Daily Course of Clusters \n')
plt.axis('off')
for i in range(1, 6):
    fig.add_subplot(rows, columns, i)
    plot_cluster('Class ' + str(i), colors[i-1])  
plt.show()
```

### Potential Service Improvements Based on Clustering Results

The presented findings from the spatio-temporal analysis show that reasons for certain pickup and return activities at bike stations are complex and diverse for both temporal and spatial influencing variables. As for analyzed characteristics of each cluster, we propose the following measures to cope with the objectives of bike sharing operation mentioned before. 


__1. Differentiation pricing integrated with incentive mechanism__
    
- This is also known as zonal pricing. Zonal pricing is appropriate virtually anywhere that returning is congested. We recommend setting prices to maintain 85-90% occupancy rates of bikes. If implemented with good user information (signs, maps and brochures that indicate returning location and price), bicycle riders can choose between more convenient but costly returning their bike, or cheaper return a short distance away.


__2. Relocating bike stations__
    
- This recommendation concerns network coverage. In general, the distribution of stations is dependent on the size and configuration of the city. In high-density cities such as Manhattan, stations densely distributed across the whole city are recommended, whereas in those cities such as Long Island and Brooklyn where density is relatively low, coverage with stations is proposed only in city centers or higher-density areas. We also suggest citibike to catch the right time to allocate enough room for people to return bikes for different clusters and supply enough bikes during the high demand hours. 


__3. Connect bike stations with public transit__
   
- A key element is the location of the stations in relation to the public transportation network. The bicycle becomes a complementary mode within the transit system because it extends the radius of influence of transit stations and stops on movements related to both access and dispersion. We strongly recommend lowering the price in zones with high residential land use to encourage residents to bike in order to solve the last mile problem for commute trips.
