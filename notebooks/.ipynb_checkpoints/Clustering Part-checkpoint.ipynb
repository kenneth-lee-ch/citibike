### Extracting Station Starts & Ends by Hour

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sqlalchemy as db
from sklearn import preprocessing
from sklearn.cluster import KMeans, Birch, SpectralBiclustering, DBSCAN
from sklearn.metrics import davies_bouldin_score, silhouette_score
```

```python
## Create database engine
sqlite_file = 'data/nyccb.sqlite'
engine = db.create_engine('sqlite:///' + sqlite_file)
```

```python
def get_OD(h1,h2):
    """
    obtain mean duration for pickups/returns
    and number of starts/ends per hour
    return a joined dataframe for data storage
    """
    hour1 = '"'+h1+'"'
    hour2 = '"'+h2+'"'
    O_hour = pd.read_sql_query('SELECT [start station name] AS Station, '
                                 '[start station id] AS ID, '
                                 'AVG(tripduration)/60 AS mean_duration_O' + h1 + h2 + 
                               ' , COUNT (*) AS O_' + h1 + h2 + 
                               ' FROM cb '
                               'WHERE strftime(\'%H\', starttime) >= ' + hour1 + 
                               ' AND strftime(\'%H\', starttime) < ' + hour2 + 
                               ' AND tripduration < 86400 '
                               'GROUP BY [start station name]', engine)

    D_hour = pd.read_sql_query('SELECT [end station name] AS Station, '
                                 '[end station id] AS ID, '
                               'AVG(tripduration)/60 AS mean_duration_D' + h1 + h2 + 
                               ', COUNT (*) AS D_' + h1 + h2 + 
                               ' FROM cb '
                               'WHERE strftime(\'%H\', stoptime) >= ' + hour1 + 
                               ' AND strftime(\'%H\', stoptime) < ' + hour2 + 
                               ' AND tripduration < 86400 '
                               'GROUP BY [end station name]', engine)
    OD_hour = pd.merge(O_hour, D_hour, on=['Station','ID'], how='outer')
    OD_hour = OD_hour.dropna()
    print(h1+h2,'\t','completed.')
    return OD_hour
```

**Warning:** The following one cell may take half an hour to run as multiple queries will be made. 

```python
OD_01 = get_OD('00','01')
OD_12 = get_OD('01','02')
OD_23 = get_OD('02','03')
OD_34 = get_OD('03','04')
OD_45 = get_OD('04','05') 
OD_56 = get_OD('05','06')
OD_67 = get_OD('06','07') 
OD_78 = get_OD('07','08') 
OD_89 = get_OD('08','09') 
OD_910 = get_OD('09','10') 
OD_1011 = get_OD('10','11') 
OD_1112 = get_OD('11','12') 
OD_1213 = get_OD('12','13') 
OD_1314 = get_OD('13','14')
OD_1415 = get_OD('14','15') 
OD_1516 = get_OD('15','16') 
OD_1617 = get_OD('16','17') 
OD_1718 = get_OD('17','18') 
OD_1819 = get_OD('18','19')
OD_1920 = get_OD('19','20')
OD_2021 = get_OD('20','21') 
OD_2122 = get_OD('21','22') 
OD_2223 = get_OD('22','23') 
OD_2324 = get_OD('23','24') 
```

```python
table = [ OD_01 , OD_12 , OD_23 , OD_34 , OD_45 , OD_56 ,
          OD_67 , OD_78 , OD_89 , OD_910 , OD_1011 , OD_1112 ,
          OD_1213 , OD_1314 , OD_1415 , OD_1516 , OD_1617 , OD_1718 ,
          OD_1819 , OD_1920 , OD_2021 , OD_2122 , OD_2223 , OD_2324 ]
OD = table[0]
for i in range(0,23):
    OD = pd.merge(OD,table[i+1],on=['Station','ID'],how='outer')

for i in np.arange(3, 3+2*48, 2):
    OD.iloc[:,i] = OD.iloc[:,i]/365 #obtain daily average pickups/returns
OD = OD.drop_duplicates('ID')
OD = OD.sort_values(by='ID')
OD = OD[OD['ID']>0]
OD
```

We can see there are some missing values because some stations may not be counted when they have no record of pickups or returns. We fill these NA values to be zero as we want to make sure the data processing will not encounter NA for calculating average duration, pickups and returns. 

```python
#calculation of total daily pickups and returns
OD = OD.fillna(0)
OD = OD.iloc[:,1:]
OD['pickup_sum'] = OD.iloc[:,np.arange(3,3+4*24,4)].sum(axis=1)
OD['return_sum'] = OD.iloc[:,np.arange(5,5+4*24,4)].sum(axis=1)
OD
```

```python
#calculation of average daily pickup-trip duration and return-trip duration
p,r=0,0
for i in range(24):
    p = p+ OD.iloc[:,2+i*4]*OD.iloc[:,3+i*4]
    r = r+ OD.iloc[:,4+i*4]*OD.iloc[:,5+i*4]
OD['pickup_duration'] = p/OD['pickup_sum']
OD['return_duration'] = r/OD['return_sum']
OD
```

```python
#calculate fration of pickups/returns
temporal = OD[['ID']]
temporal = temporal.merge(OD.iloc[:,-4:],left_index=True,right_index=True)
for i in range(0,24):
    temporal['pickup'+str(i)] = OD.iloc[:,3+i*4]/temporal['pickup_sum']
    temporal['return'+str(i)] = OD.iloc[:,5+i*4]/temporal['return_sum']
temporal
```

So far, we have already extracted temporal data for clustering, and then we will combine it with external spatial data. 

```python
#combine with spatial data extracted by ArcGIS (external)
spatial = pd.read_csv('data/spatial.csv').drop_duplicates('ID')
cluster = temporal.merge(right=spatial,on='ID',how='inner')
cluster
```

For now, the data for clustering have benn extracted completely. 

**Standardization & weighting of variables**

```python
##standardization of variables
#station_list = pd.read_csv('spatial_data/station_list.csv')
X = cluster
X = X.iloc[:,1:]
#X = X.iloc[:,:]
x0 = X.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x0)
X = pd.DataFrame(x_scaled)
X
```

```python
#giving weights for temporal and spatial variables for weight balancing
X1 = X.iloc[:,0:2]*4 #biking activities
X2 = X.iloc[:,2:4]*3.5 #duration
X2 = X.iloc[:,4:52]*1 #temporal
X3 = X.iloc[:,52:]*2.5 #spatial
Y = X1.merge(right=X2,left_index=True,right_index=True)\
.merge(right=X3,left_index=True,right_index=True)
Y
```

We are trying three different clustering algorithms and to make decision on which algorithm and the number of clusters to use finally. 

```python
dbs=[]
for i in range(3,11):
    c1 = KMeans(n_clusters=i, random_state=1, algorithm='auto').fit(Y) 
    label1 = c1.labels_
    c2 = Birch(n_clusters=i).fit(Y)
    label2 = c2.labels_
    c3 = SpectralBiclustering(n_clusters=i, random_state=1).fit(Y)
    label3 = c3.row_labels_

    dbs.append([silhouette_score(Y, label1),
                silhouette_score(Y, label2),
                silhouette_score(Y, label3)])
cluster_validation_sil = pd.DataFrame(dbs).rename(columns={0:'K-Means',
                                              1:'Birch',
                                              2:'Spectral Biclustering'})
```

```python
dbs=[]
for i in range(3,11):
    c1 = KMeans(n_clusters=i, random_state=1, algorithm='auto').fit(Y) 
    label1 = c1.labels_
    c2 = Birch(n_clusters=i).fit(Y)
    label2 = c2.labels_
    c3 = SpectralBiclustering(n_clusters=i, random_state=1).fit(Y)
    label3 = c3.row_labels_

    dbs.append([davies_bouldin_score(Y, label1),
                davies_bouldin_score(Y, label2),
                davies_bouldin_score(Y, label3)])
cluster_validation_dav = pd.DataFrame(dbs).rename(columns={0:'K-Means',
                                              1:'Birch',
                                              2:'Spectral Biclustering'})
```

```python
plt.rc('font', family='arial', size=12)
cluster_validation_sil.iloc[:,:3].plot(kind='line',linewidth=2,
                                   alpha=0.7,rot=1,figsize=(8,4),legend=True)
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
#plt.yticks(np.arange(0,10000,2000))
plt.xticks(np.arange(0,8,1),np.arange(3,11,1))
plt.grid(axis="y",linestyle='--',linewidth=0.5)
plt.show()
```

```python
plt.rc('font', family='arial', size=12)
cluster_validation_dav.iloc[:,:3].plot(kind='line',linewidth=2,linestyle='--',
                                   alpha=0.7,rot=1,figsize=(8,4),legend=True)
plt.xlabel('Number of Clusters')
plt.ylabel('Davies Bouldin Score')
#plt.yticks(np.arange(0,10000,2000))
plt.xticks(np.arange(0,8,1),np.arange(3,11,1))
plt.grid(axis="y",linestyle='--',linewidth=0.5)
plt.show()
```

We choose the number of clusters to be 6 and use K-Means algorithm to do the following clustering.

```python
#clustering into 6 clusters
c = KMeans(n_clusters=6).fit(Y)
print(silhouette_score(Y, c.labels_))
result = cluster
result['label'] = c.labels_
result['label'] = result['label'].replace({0:'Class 1',
                              1:'Class 2',
                              2:'Class 3',
                              3:'Class 4',
                              4:'Class 5',
                              5:'Class 6'})
station_list = pd.read_csv('spatial_data/station_list.csv')
result = pd.merge(result, station_list, on=['ID'], how='left')
group_result = result.groupby('label')
print(group_result.size())
```

```python
#save the result to csv
result.to_csv('data/result.csv')
```
