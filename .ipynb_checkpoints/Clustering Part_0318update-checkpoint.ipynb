### ii. Unsupervised Learning - Clustering

Having the exploratory findings in the previous part about NYC Citi Bike system, we are more interested in whether there can be certain types of bike stations where their biking activities are significantly different. These differences are directly shown in the average daily activities of pickups or returns of bicycles at stations, which can be seen both temporally and spatially. For example, some bike stations that are located in downtown business areas where a lot of companies, industries and  headquarters generate numerous employment may receive high and concentrated volume of bike returns in the weekday morning peak period. This can also be true that bike stations located around the subway lines may encounter intensive pickup demands at a certain time whereas the capacity of these bike stations can not support. On the contrary, a large portion of bikes in the bike sharing system may not be in efficient use and thus face a long time with empty occupancy. 

Taking the standpoints of the operator of bike sharing systems, and the planners of public and private sectors, it is necessary and essential for them to come up with __a comprehensive framework of bike station classification integrating multi data resources__. Here, __cluster analysis__ is used in order to group stations according to their average daily bike pickup and return activities. The anticipated outcome of clustering in the form of spatial and temporal patterns will be thoroughly examined to discover location dependent reasons for biking activity patterns. 

The examination criteria of clusters in terms of cohesion and separation of data points, which is also considered as dissimilarity among clusters. In particular, Euclidean distances that can be computed for large data sets are used to quantify the dissimilarity. On the basis of this concept, Silhouette score and Davies-Bouldin score are proposed to provide a succinct graphical representation of how well each cluster has been classified. The calculation of two scores are well defined in the scikit-learn package in python that is intensively used in the previous part of data preprocessing.

```python
#generate series of hours from 0 to 24
hour = ["%.2d" % i for i in range(25)]
#initialize empty dataframe for merging
#O for origin(pickup), D for destination(return)
OD = pd.DataFrame(columns = ['Station','ID'])
#obtain mean durations for p/r and number of p/r by hour and combine it to a complete dataframe
for i in range (24):
    OD_hour = get_OD(hour[i],hour[i+1])
    OD = pd.merge(OD,OD_hour,on=['Station','ID'],how='outer')

#obtain daily average pickups/returns
for i in np.arange(3, 3+2*48, 2):
    OD.iloc[:,i] = OD.iloc[:,i]/365 #365: converting total annual value to daily

##clean the OD data
#ID should be positive and unique
OD = OD.drop_duplicates('ID')
OD = OD[OD['ID']>0]
#sort ID ascending
OD = OD.sort_values(by='ID')
#avoid missing values when calculating
OD = OD.fillna(0)

#calculation of total daily pickups and returns
OD['pickup_sum'] = OD.iloc[:,np.arange(3,3+4*24,4)].sum(axis=1)
OD['return_sum'] = OD.iloc[:,np.arange(5,5+4*24,4)].sum(axis=1)

#calculation of average daily pickup-trip duration and return-trip duration
p,r=0,0
for i in range(24):
    #weighted daily pickup sumation = mean duration of pickup * number of pickups
    p = p+ OD.iloc[:,2+i*4]*OD.iloc[:,3+i*4]
    #weighted daily return sumation = mean duration of return * number of return
    r = r+ OD.iloc[:,4+i*4]*OD.iloc[:,5+i*4]
#average daily pickup-trip duration and return-trip duration
OD['pickup_duration'] = p/OD['pickup_sum']
OD['return_duration'] = r/OD['return_sum']
OD
```

```python
#calculate fration of pickups/returns
temporal = OD[['ID']]
#obtain last 4 columns of variables at the beginning of dataframe
temporal = temporal.merge(OD.iloc[:,-4:],left_index=True,right_index=True)
#calculate fraction of pickups/returns by hour
for i in range(0,24):
    temporal['pickup'+str(i)] = OD.iloc[:,3+i*4]/temporal['pickup_sum']
    temporal['return'+str(i)] = OD.iloc[:,5+i*4]/temporal['return_sum']

#import spatial data extracted by ArcGIS (external)
spatial = pd.read_csv('data/spatial.csv').drop_duplicates('ID')

#combine into final dataset for clustering
cluster = temporal.merge(right=spatial,on='ID',how='inner')
cluster
```

For now, the data for clustering have been extracted completely. 

**Standardization & weighting of variables**

```python
def standardize(df):
    """
    Conduct standardization of da dataframe. 
    Argument: a pandas dataframe without standardization
    Return: a standardized dataframe
    """
    
    x0 = df.values      #returns a numpy array
    min_max_scaler = preprocessing.MinMaxScaler()
    x_scaled = min_max_scaler.fit_transform(x0)
    df_s = pd.DataFrame(x_scaled)
    
    return df_s
```

```python
##standardization of variables
X = standardize(cluster.iloc[:,1:])  #neglect the first ID column

##giving weights for temporal and spatial variables 
##for weight balancing (too many temporal variables)

#first part
X1 = X.iloc[:,0:2]*4 #biking activities
#second part
X2 = X.iloc[:,2:4]*3.5 #mean duration
#third part
X3 = X.iloc[:,4:52]*1 #temporal
#forth part
X4 = X.iloc[:,52:]*5 #spatial

#combine to weighted data
Y = X1.merge(right=X2,left_index=True,right_index=True)\
.merge(right=X3,left_index=True,right_index=True)\
.merge(right=X4,left_index=True,right_index=True)
Y
```

- As a summary, there are 69 variables in the extracted dataset for clustering, which are standardized for the ongoing cluster analysis. 
- We are trying three different clustering algorithms and to make decision on which algorithm and the number of clusters to use finally. 
- The metric we adopt to measure the performance is SIlhouette Index and Davies-Bouldin Index, which represent the cohesion and seperation of data points. 
- k-means, Birch (balanced iterative reducing and clustering using hierarchies) and Spectral Biclustering (of data dimension reduction methods) are applied for cluster analysis. KM is one of the most well-known partitional cluster algorithms.

__The following analysis is written based on the order of clusters. The order may change if the following cell is re-run and thus the analysis of each cluster may be wrong in order.__

```python
#clustering into 5 clusters
c = KMeans(n_clusters=5).fit(Y)
print(silhouette_score(Y, c.labels_))
#append cluster label result to original data
result = cluster
result['label'] = c.labels_
#rename clusters
result['label'] = result['label'].replace({0:'Class 1',
                              1:'Class 2',
                              2:'Class 3',
                              3:'Class 4',
                              4:'Class 5',
                              5:'Class 6'})
#read in station list with lat and lon which is saved by making query in SQL in the previous part
station_list = pd.read_csv('spatial_data/station_list.csv')
station_list = station_list.drop_duplicates('ID')
#merge cluster result with lat and lon of stations
result = pd.merge(result, station_list, on=['ID'], how='left')
group_result = result.groupby('label')
#see cluster size
print(group_result.size()) 
```

```python
#save the result to csv
result.to_csv('data/result.csv')
```

The K-Means clustering algorithm for five clusters is finally implemented by the indication of Silhouette and Davies-Bouldin scores and thus five clusters of bike stations in New York city are obtained. Regarding the activity at stations, the mean values which can be conceived as the centroids of each cluster can be validated both temporally and spatially. In a summary, the numbers of stations assigned to each cluster are 155, 261, 121, 50 and 345, from Class 1 to Class 5 clusters respectively. The consequent sample sizes of clusters are preferable, which also meet the prespecified clustering examination criteria.

```python
#plotting fractions of p/r by hour for the next three cells
_pickup = group_result.mean().iloc[:, [i for i in range(5,5+48,2)]].T
_return = group_result.mean().iloc[:, [i for i in range(6,6+48,2)]].T
```

```python
def plot_cluster(cluster, color):
    """
    plot the hourly fractions of pickups and returns of each cluster
    Arguments: 
        cluster: name of the cluster predefined, str
        color: six-digit color series name, str 
    """
    c = pd.DataFrame(_pickup['Class 1'],_return['Class 1'])
    plt.rc('font', family='arial', size=12)
    _pickup[cluster].plot(kind='line',linestyle='-',color=color, label=cluster + ' pickup')
    _return[cluster].plot(kind='line',linestyle='--',color=color, label=cluster + ' return')
    plt.xlabel('Hour of Day')
    plt.ylabel('Fraction of Daily Activities (%)')
    plt.legend()
    plt.yticks(np.arange(0,0.2,0.04),np.arange(0,20.0,4.0))
    plt.xticks(np.arange(0,30,6),np.arange(0,30,6))
    plt.grid(axis="y",linestyle='--',linewidth=0.5)
    
    return
```

```python
fig=plt.figure(figsize=(17,8))
columns = 3
rows = 2
colors = ['#3498db','#1abc9c','#f1c40f','#e67e22','#e74c3c']

plt.title('Pickups & Returns in the Daily Course of Clusters \n')
plt.axis('off')
for i in range(1, 6):
    fig.add_subplot(rows, columns, i)
    plot_cluster('Class ' + str(i), colors[i-1])  
plt.show()
```

### Potential Improvements Based on Clustering Results

The presented findings from the spatio-temporal analysis show that reasons for certain pickup and return activities at bike stations are complex and diverse for both temporal and spatial influencing variables. As for analyzed characteristics of each cluster, we propose the following measures to cope with the objectives of bike sharing operation mentioned before. 


__1. Differentiation pricing integrated with incentive mechanism__
    
- This is also known as zonal pricing. Zonal pricing is appropriate virtually anywhere that returning is congested. We recommend setting prices to maintain 85-90% occupancy rates of bikes. If implemented with good user information (signs, maps and brochures that indicate returning location and price), bicycle riders can choose between more convenient but costly returning their bike, or cheaper return a short distance away.


__2. Relocating bike stations__
    
- This recommendation concerns network coverage. In general, the distribution of stations is dependent on the size and configuration of the city. In high-density cities such as Manhattan, stations densely distributed across the whole city are recommended, whereas in those cities such as Long Island and Brooklyn where density is relatively low, coverage with stations is proposed only in city centers or higher-density areas.


__3. Connect bike stations with public transit__
   
- A key element is the location of the stations in relation to the public transportation network. The bicycle becomes a complementary mode within the transit system because it extends the radius of influence of transit stations and stops on movements related to both access and dispersion. We strongly recommend lowering the price in zones with high residential land use to encourage residents to bike in order to solve the last mile problem for commute trips.
